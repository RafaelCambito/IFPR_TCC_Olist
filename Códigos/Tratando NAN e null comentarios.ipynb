```python
#Importando biblioteca pandas para auxiliar no processo de análise e manipulação dos dados
import pandas as pd
from google.colab import files
```

```python
df = pd.read_csv('pedido_comentarios.csv')
```

```python
# Verificar se existem valores nulos
valores_nulos = df.isnull().sum()

# Mostrar a quantidade de valores nulos por coluna
print(valores_nulos)

# Verificar se há alguma coluna com valores NaN
tem_nan = df.isna().sum().sum() > 0

print(f"O DataFrame contém valores NaN: {tem_nan}")
```

```python
# Mostrar todas as linhas que têm pelo menos um valor NaN
linhas_com_nan = df[df.isna().any(axis=1)]
print(linhas_com_nan)
```

```python
# Substituir valores NaN por "sem registro"
df.fillna("sem registro", inplace=True)

# Verificar se a substituição foi bem-sucedida
print(df.head())  # Mostrar as primeiras linhas do DataFrame após a substituição
```

```python
# Verificar se existem valores nulos
valores_nulos = df.isnull().sum()

# Mostrar a quantidade de valores nulos por coluna
print(valores_nulos)

# Verificar se há alguma coluna com valores NaN
tem_nan = df.isna().sum().sum() > 0

print(f"O DataFrame contém valores NaN: {tem_nan}")
```

```python
# Salvar o DataFrame atualizado em um novo arquivo CSV
df.to_csv('arquivo_tratado.csv', index=False)

```
